```python
#importing the beautiful soup
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title1=soup.find_all('h4',class_="italic")
title1
```

```python
Book_Name=[]
for i in title1:
    Book_Name.append(i.text.replace('\n',' '))
Book_Name
```

```python
title2=soup.find_all('p',class_="sans bold")
title2
```

```python
Author_Name=[]
for i in title2:
    Author_Name.append(i.text.replace('\n',' '))
Author_Name
```

```python
title3=soup.find_all('p',class_="genre-links hidden-phone")
title3
```

```python
Genere=[]
for i in title3:
    Genere.append(i.text.replace('\n',' '))
Genere
```

```python
title4=soup.find_all('p',class_="excerpt")
title4
```

```python
Review=[]
for i in title4:
    Review.append(i.text.replace('\n',' '))
Review
```

```python
import pandas as pd
Book_Review=pd.DataFrame({})
Book_Review['Book_Name']=Book_Name
Book_Review['Author']=Author_Name
Book_Review['Denere']=Genere
Book_Review['Review']=Review
Book_Review
```
