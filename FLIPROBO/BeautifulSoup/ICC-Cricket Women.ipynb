```python
#importing the beautiful soup
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://crickettimes.com/icc-rankings/women-odi-team-rankings/")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title2=soup.find_all('div',class_="team-ranking-row")
ICC_Women=[]
for i in title2:
    ICC_Women.append(i.text.replace('\n',' '))
print(ICC_Women)
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi")
page
```

```python
page.content
```

```python
soup1=BeautifulSoup(page.content)
soup1
```

```python
title2=soup1.find_all('tr',class_="table-body")
ICC_Batsmen=[]
for i in title2:
    ICC_Batsmen.append(i.text.replace('\n',' '))
print(ICC_Batsmen)
```
