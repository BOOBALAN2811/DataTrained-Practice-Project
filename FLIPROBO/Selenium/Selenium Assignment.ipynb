```python
driver=webdriver.Chrome("chromedriver.exe")
url = 'https://www.flipkart.com/'
driver.get(url)
```

```python
DAJ=driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input")
DAJ.send_keys("sunglasses")
DAB=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
DAB.click()
```

```python
import time
Brand=[]
Description=[]
Price=[]
Discount=[]
for i in range(0,3):
    for j in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
        Brand.append(j.text)
    for K in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
        Description.append(K.text)
    for L in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
        Price.append(L.text)
    for M in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']//span"):
        Discount.append(M.text)
    driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
    time.sleep(10)
import pandas as pd
Sunglass=pd.DataFrame({})
Sunglass['Brand']=Brand
Sunglass['Description']=Description
Sunglass['Price']=Price
Sunglass['Discount']=Discount
Sunglass
```
