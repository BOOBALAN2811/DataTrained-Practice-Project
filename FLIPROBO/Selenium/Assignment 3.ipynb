```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

Q1 - Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.

Q2 - In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a dataframe and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”.

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)
url = "https://www.amazon.in/"
driver.get(url)
time.sleep(6)
Input = input('Enter a product here: ')
search_bar=driver.find_element_by_id("twotabsearchtextbox")
search_bar.clear()
search_bar.send_keys(Input)
time.sleep(5)
search_button=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']/span/input")
search_button.click()
```

```python
Brand_Name=[]
Product_Name=[]
Rating=[]
No_of_Ratings=[]
Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]
Product_URL=[]
```

```python
for i in range(0,3):
    for a in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']/span"):
        if a.text is None :
            Product_Name.append("--")      
        else:
            Product_Name.append(a.text)
    for d in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
        Product_URL.append(d.get_attribute('href'))  
    driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    time.sleep(5)
```

```python
for url in Product_URL:
    driver.get(url)
    try:         
        rating=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append("NO rating")
        
for url1 in Product_URL:
    driver.get(url1)
    try:
        price=driver.find_element_by_xpath("//td[@class='a-span12']//span")
        Price.append(price.text)
    except NoSuchElementException:
        Price.append("NO Price")

for url2 in Product_URL:
    driver.get(url2)
    try:
        Brand=driver.find_element_by_xpath("//td[@class='a-span9']//span")
        Brand_Name.append(Brand.text)
    except NoSuchElementException:
        Brand_Name.append("NO Brand Name") 

for url3 in Product_URL:
    driver.get(url3)
    try:
        number=driver.find_element_by_id("acrCustomerReviewText")
        No_of_Ratings.append(number.text)
    except NoSuchElementException:
        No_of_Ratings.append("NO Rating")
        
for url4 in Product_URL:
    driver.get(url4)
    try:
        AS=driver.find_element_by_id("availability")
        Availability.append(AS.text)
    except NoSuchElementException:
        Availability.append("NO Details")
        
for url5 in Product_URL:
    driver.get(url5)
    try:
        A=driver.find_element_by_xpath("dic[@class='a-section a-spacing-top-mini']/span/b")
        Expected_Delivery.append(A.text)
    except NoSuchElementException:
        Expected_Delivery.append("NO Details")
        
for url6 in Product_URL:
    driver.get(url6)
    try:
        B=driver.find_element_by_id("productSupportAndReturnPolicy-return-policy-anchor-text")
        Return_Exchange.append(B.text)
    except NoSuchElementException:
        Return_Exchange.append("NO Details")
```

```python
Mobile=pd.DataFrame({})
Mobile['Brand_Name']=Brand_Name
Mobile['Product_Name']=Product_Name
Mobile['Rating']=Rating
Mobile['No_of_Ratings']=No_of_Ratings
Mobile['Price']=Price
Mobile['Return_Exchange']=Return_Exchange
Mobile['Expected_Delivery']=Expected_Delivery
Mobile['Availability']=Availability
Mobile['Product_URL']=Product_URL
Mobile
```

```python
Mobile.to_csv('Iphone_Details.csv')
```

Write a python program to access the search bar and search button on images.google.com and
scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)
url = "https://www.digit.in/top-products/best-gaming-laptops-40.html"
driver.get(url)
time.sleep(6)
```

```python
URL=[]
for i in driver.find_elements_by_xpath("//div[@class='TopNumbeHeading sticky-footer']/a"):
    URL.append(i.get_attribute('href'))
```

```python
Name=[]
Status=[]
Price_Details=[]
Digit_Rating=[]
```

```python
for url in URL:
    driver.get(url)
    try:         
        rating=driver.find_element_by_xpath("//div[@class='heading-wraper']//h1")
        Name.append(rating.text)
    except NoSuchElementException:
        Name.append("NO Name")
        
for url1 in URL:
    driver.get(url1)
    try:
        price=driver.find_element_by_xpath("//div[@class='Block-status']//b")
        Status.append(price.text)
    except NoSuchElementException:
        Status.append("NO Status")

for url2 in URL:
    driver.get(url2)
    try:
        Brand=driver.find_element_by_xpath("//div[@class='Onward-price']/div/b")
        Price_Details.append(Brand.text)
    except NoSuchElementException:
        Price_Details.append("NO Price") 

for url3 in URL:
    driver.get(url3)
    try:
        number=driver.find_element_by_css_selector("p.yellowtext")
        Digit_Rating.append(number.text)
    except NoSuchElementException:
        Digit_Rating.append("NO Rating")
```

```python
Laptop=pd.DataFrame({})
Laptop['Name']=Name
Laptop['Status']=Status
Laptop['Price_Details']=Price_Details
Laptop['Digit_Rating']=Digit_Rating
Laptop
```
